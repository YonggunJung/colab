```python
import pandas as pd
import numpy as np
from collections import Counter
from google.colab import files
```

```python
# CSV 업로드
# uploaded = files.upload()
import io
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/로또/data/jlotto.csv', header=None)
df
```

```python
# 번호 데이터만 추출 (첫 행 제외)
lotto_data = df.values[:]
lotto_data = [list(map(int, row)) for row in lotto_data]
lotto_data
```

```python
# 빈도수 행렬 초기화 (45x45)
# freq_matrix[i][j]는 숫자 i+1이 포함된 행 다음 줄에서 숫자 j+1이 몇 번 나왔는지
freq_matrix = np.zeros((45, 45), dtype=int)
freq_matrix
```

```python
# 빈도수 행렬 계산
for i in range(len(lotto_data) - 1):  # 마지막 줄은 다음 줄이 없어서 제외
    current_row = lotto_data[i]
    next_row = lotto_data[i + 1]

    for num in current_row:
        for next_num in next_row:
            freq_matrix[num - 1][next_num - 1] += 1
freq_matrix
```
