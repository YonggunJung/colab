```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/로또/data/pension.csv')
df
```

```python
Xtrain = df.drop('pension', axis = 1)
Xtrain
```

```python
Xtest = np.array(Xtrain.iloc[-1])
Xtest
```

```python
Xtrain.drop(Xtrain.index[-1], inplace = True)
Xtrain
```

```python
ytrain = df.drop('win', axis = 1)
ytrain
```

```python
ytrain.dropna(inplace = True)
ytrain
```

```python
ytrain =ytrain.astype('int64')
ytrain
```

```python
from sklearn.model_selection import train_test_split

Xtr, Xval, ytr, yval = train_test_split(Xtrain, ytrain,
                                        test_size = 0.3,
                                        random_state = 20242)
```

```python
# 머신러닝 알고리즘 분류
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import VotingClassifier
from lightgbm import LGBMClassifier
from xgboost import XGBClassifier

model1 = KNeighborsClassifier(n_neighbors = 6)
model2 = LogisticRegression(random_state = 20242)
model3 = SVC(random_state = 20242, probability = True)
model4 = DecisionTreeClassifier(random_state = 20242,
                                max_depth = 50)
model5 = RandomForestClassifier(random_state = 20242,
                                max_depth = 2,
                                n_estimators= 5)
model6 = AdaBoostClassifier(random_state = 20242)
model7 = GradientBoostingClassifier(random_state = 20242)
model8 = LGBMClassifier(random_state = 20242)
# model9 = XGBClassifier(random_state = 20242, silent = True, verbosity = 0)

model_list1 = [
    ('m1', KNeighborsClassifier()),
    ('m2', LogisticRegression(random_state = 20242)),
    ('m3', SVC(random_state = 20242, probability = True)),
    ('m4', DecisionTreeClassifier(random_state = 20242)),
    ('m5', RandomForestClassifier(random_state = 20242)),
    ('m6', AdaBoostClassifier(random_state = 20242)),
    ('m7', GradientBoostingClassifier(random_state = 20242)),
    ('m8', LGBMClassifier(random_state = 20242))
    # ('m9', XGBClassifier(random_state = 20242, silent = True, verbosity = 0))
]

# 배깅 운영 방식은 위에서 사용한 앙상블, 부스팅들이 사용하고 있다.
# 보팅 운영 방식도 테스트 해보기 위해 사용
# 매개변수 voting은 투표 방식 설정 (hard(다수결)/soft(확률))

model10 = VotingClassifier(estimators=model_list1, voting='hard')

model_list2 = [
    ('m10', KNeighborsClassifier()),
    ('m20', LogisticRegression(random_state = 20242)),
    ('m30', SVC(random_state = 20242, probability = True)),
    ('m40', DecisionTreeClassifier(random_state = 20242)),
    ('m50', RandomForestClassifier(random_state = 20242)),
    ('m60', AdaBoostClassifier(random_state = 20242)),
    ('m70', GradientBoostingClassifier(random_state = 20242)),
    ('m80', LGBMClassifier(random_state = 20242))
    # ('m90', XGBClassifier(random_state = 20242, silent = True, verbosity = 0))
]

model11 = VotingClassifier(estimators=model_list1, voting='soft')
```

```python
model1.fit(Xtr, ytr)
model4.fit(Xtr, ytr)
model5.fit(Xtr, ytr)

model2.fit(Xtr, ytr)
model3.fit(Xtr, ytr)
model6.fit(Xtr, ytr)
model7.fit(Xtr, ytr)
model8.fit(Xtr, ytr)
# model9.fit(Xtr, ytr)
model10.fit(Xtr, ytr)
model11.fit(Xtr, ytr)


pred1 = model1.predict(Xval)
pred4 = model4.predict(Xval)
pred5 = model5.predict(Xval)

pred2 = model2.predict(Xval)
pred3 = model3.predict(Xval)
pred6 = model6.predict(Xval)
pred7 = model7.predict(Xval)
pred8 = model8.predict(Xval)
# pred9 = model9.predict(Xval)
pred10 = model10.predict(Xval)
pred11 = model11.predict(Xval)

print('KNN : ', pred1)
print('DecisionTree : ', + pred4)
print('RandomForest : ', + pred5)

# print('LogisticRegression' + pred2)
# print('SVM' + pred3)
# print('AdaBoost' + pred6)
# print('GradientBoost' + pred7)
# print('LGBM' + pred8)
# print('XGBoost' + pred9)
# print('Voting(hard)' + pred10)
# print('Voting(soft)' + pred11)
```
