```python
# 사이킷런 제공 토이 데이터셋
# 붓꽃 데이터셋
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris
```

```python
# 컬럼=속성=피쳐 확인하기
# iris['feature_names']  이거도 가능
iris.feature_names
```

```python
# 라벨=y=정답지 확인하기
#iris['target]
print(len(iris.target))
iris.target
```

```python
# 라벨명 확인
# iris['target_names']
iris.target_names
```

```python
import pandas as pd
import numpy as np
irisdf = pd.DataFrame(iris.data, columns = iris.feature_names)
irisdf
```

```python
# 라벨 컬럼 추가
irisdf['label'] = iris.target
irisdf
```

```python
# 데이터 정보 확인
irisdf.info()
```

```python
# 시각화
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns

# 그래프 기본 사이즈 설정
plt.rcParams['figure.figsize'] = [8, 4.5]

print(iris.target_names)
sns.scatterplot(x = 'sepal length (cm)', y = 'sepal width (cm)', data = irisdf, hue = 'label')
```
