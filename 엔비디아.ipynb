```python
# 미국 주식: 티커를 사용. 예를 들어, 'AAPL'(애플), 'AMZN'(아마존), 'GOOG'(구글)
# 엔비디아 가져오기
# 데이터 프레임: 날짜, 시작가, 고가, 저가, 종가, 거래량, 전일 대비 변화랑
nv = fdr.DataReader('NVDA')
del nv['Open'], nv['High'], nv['Low'], nv['Volume'], nv['Adj Close']
nv.columns = ['엔비디아']
nv
```

```python
# 국내 주식: 단축 코드(6자리)를 사용. 예를 들어, '005930'(삼성전자), '215600'(신라젠)
# MDS테크 :        086960 코스피
mds = fdr.DataReader('086960')
del mds['Open'], mds['High'], mds['Low'], mds['Volume'], mds['Change']
mds.columns = ['MDS테크']
mds
```

```python
# 씨이랩 :         189330 코스피
ce = fdr.DataReader('189330')
del ce['Open'], ce['High'], ce['Low'], ce['Volume'], ce['Change']
ce.columns = ['씨이랩']
ce
```

```python
# 아이티아이즈 :   372800 코스피
ite = fdr.DataReader('372800')
del ite['Open'], ite['High'], ite['Low'], ite['Volume'], ite['Change']
ite.columns = ['아이티아이즈']
ite
```

```python
# 이수페타시스 :   007660 코스피
esu = fdr.DataReader('007660')
del esu['Open'], esu['High'], esu['Low'], esu['Volume'], esu['Change']
esu.columns = ['이수페타시스']
esu
```

```python
# 콤텍시스템 :     031820 코스피
cts = fdr.DataReader('031820')
del cts['Open'], cts['High'], cts['Low'], cts['Volume'], cts['Change']
cts.columns = ['콤텍시스템']
cts
```

```python
# 제이씨현시스템 : 033320 코스피
jc = fdr.DataReader('033320')
del jc['Open'], jc['High'], jc['Low'], jc['Volume'], jc['Change']
jc.columns = ['제이씨현시스템']
jc
```

```python
# 아이크래프트 :   052460 코스피
icft = fdr.DataReader('052460')
del icft['Open'], icft['High'], icft['Low'], icft['Volume'], icft['Change']
icft.columns = ['아이크래프트']
icft
```

```python
# 세코닉스 :       053450 코스피

scnix = fdr.DataReader('053450')
del scnix['Open'], scnix['High'], scnix['Low'], scnix['Volume'], scnix['Change']
scnix.columns = ['세코닉스']
scnix
```

```python
# 한미반도체 :       042700 코스피

hm = fdr.DataReader('042700')
del hm['Open'], hm['High'], hm['Low'], hm['Volume'], hm['Change']
hm.columns = ['한미반도체']
hm
```

```python
# mds, ce, ite, esu, cts, jc, icft, scnix
nvs = nv.merge(mds, left_on='Date', right_on = 'Date').merge(ce, left_on='Date', right_on = 'Date')\
              .merge(ite, left_on='Date', right_on = 'Date').merge(esu, left_on='Date', right_on = 'Date')\
              .merge(cts, left_on='Date', right_on = 'Date').merge(jc, left_on='Date', right_on = 'Date')\
              .merge(icft, left_on='Date', right_on = 'Date').merge(scnix, left_on='Date', right_on = 'Date')\
              .merge(hm, left_on='Date', right_on = 'Date')
nvs
```

```python
# 피어슨 상관 관계 확인
nvsc = nvs.corr()
nvsc
```

```python
import seaborn as sns
# 히트맵으로 확인
sns.heatmap(data = nvsc ,annot = True, fmt='.2f', cmap='Greens')
plt.title('엔비디아 관련주 상관관계')
# plt.xticks(rotation=25)
# plt.savefig('data/Image/코스피상관관계월별.png')
plt.show()
```

```python
# 21년 11월 부터 보았을때 한미반도체와 이수페타시스 이 두 종목이 90%이상 강한 양의 상관 관계가 보인다
# 이 종목으로 조금더 알아 보자
```

```python
# 2020년도 부터 확인해보자 정확한 기사는 못 찾았지만
# 기사들로 미루어 볼때 이정도 인거 같음
nv1 = fdr.DataReader('NVDA', '2020')
del nv1['Open'], nv1['High'], nv1['Low'], nv1['Volume'], nv1['Adj Close']
nv1.columns = ['엔비디아']
nv1
```

```python
# 한미반도체 :       042700 코스피

hm1 = fdr.DataReader('042700', '2020')
del hm1['Open'], hm1['High'], hm1['Low'], hm1['Volume'], hm1['Change']
hm1.columns = ['한미반도체']
hm1
```

```python
# 이수페타시스 :   007660 코스피

esu1 = fdr.DataReader('007660', '2020')
del esu1['Open'], esu1['High'], esu1['Low'], esu1['Volume'], esu1['Change']
esu1.columns = ['이수페타시스']
esu1
```

```python
nvs1 = nv1.merge(esu1, left_on = 'Date', right_on = 'Date').merge(hm1, left_on = 'Date', right_on = 'Date')
nvs1
```

```python
nvsc1 = nvs1.corr()
nvsc1
```

```python
# 히트맵
sns.heatmap(data = nvsc1, annot = True, fmt = '.2f', cmap = 'Greens')
plt.title('엔비디아, 이수페타시스, 한미반도체 상관관계')
plt.xticks(rotation = 20)  # xticks = x축 글자들, rotation = 회전
# plt.savefig('data/Image/코스피상관관계월별.png')  # 저장하는거
plt.show()
# 4년동안 아주 강한 양의 상관관계를 보인다.
# 요즘 동향 뉴스를 알아 볼것 요망
```
