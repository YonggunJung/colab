```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/로또/data/lotto2.csv')
df
```

```python
Xtrain = df.drop(['일', '이', '삼', '사','오', '육'], axis = 1)
Xtrain
```

```python
Xtest = Xtrain.iloc[Xtrain.index[-1]]
Xtest
```

```python
Xtrain.drop(Xtrain.index[-1], inplace = True)
Xtrain
```

```python
Xtrain = pd.get_dummies(Xtrain)
Xtest = pd.get_dummies(Xtest)

Xtrain
```

```python
ytrain = df.drop(['1', '2', '3', '4','5', '6'], axis = 1)
ytrain
```

```python
ytrain.dropna(inplace = True)
ytrain
```

```python
ytrain =ytrain.astype(int)
ytrain
```

```python
from sklearn.model_selection import train_test_split

Xtr, Xval, ytr, yval = train_test_split(Xtrain, ytrain,
                                        test_size = 0.15,
                                        random_state = 20241)
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import StratifiedGroupKFold

# KFold 생성
kfold = KFold(n_splits = 13, shuffle = True, random_state = 20241)
```

```python

# 머신러닝 알고리즘 - 회귀용
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.linear_model import ElasticNet
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import GradientBoostingRegressor
from lightgbm import LGBMRegressor
from xgboost import XGBRegressor
from sklearn.ensemble import VotingRegressor

model1 = KNeighborsRegressor(n_neighbors = 350,
                              leaf_size = 6)
model2 = LinearRegression()
model3 = Ridge(random_state = 20241)
model4 = Lasso(random_state = 20241)
model5 = ElasticNet(random_state = 20241)
model6 = SVR()
model7 = DecisionTreeRegressor(random_state = 20241,
                                max_depth = 50)
model8 = RandomForestRegressor(random_state = 20241,
                                max_depth = 2,
                                n_estimators= 60)
model9 = AdaBoostRegressor(random_state = 20241)
model10 = GradientBoostingRegressor(random_state = 20241)
model11 = LGBMRegressor(random_state = 20241)
model12 = XGBRegressor(random_state = 20241, silent = True, verbosity = 0)
```

```python

# 교차 검증용
# 지표를 하나만 설정할 경우
from sklearn.model_selection import cross_val_score
# 지표를 두개 이상 설정할 경우
from sklearn.model_selection import cross_validate

# 평가함수 - 회귀용
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

# 교차 검증
try:
  for idx, ml in enumerate(check_list):
      score = cross_val_score(ml, Xtr, ytr, scoring='neg_mean_squared_error', cv=kfold)


  #     현재 알고리즘의 이름과 평가 점수의 평균을 출력
      print(f'{check_name_list[idx]} : {score.mean()}')
  #     현재 알고리즘의 이름과 평가 점수의 평균을 리스트에 담는다
      score_list.append(score.mean())
      model_name_list.append(check_name_list[idx])
      best_model_list.append(ml)
except :
  pass
```
