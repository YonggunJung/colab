```python
import pandas as pd
import numpy as np
```

```python
from sklearn.datasets import load_wine
wine = load_wine()
wine
```

```python
dfwine = pd.DataFrame(wine.data, columns = wine.feature_names)
dfwine
```

```python
dfwine['label'] = wine.target
dfwine
```

```python
dfwine.info()
```

```python
# 라벨 종류 확인
dfwine['label'].unique()
```

```python
# 라벨 분포 확인
dfwine['label'].value_counts()
```

```python
# 상관관계 확인
cor = dfwine.corr()
cor
```

```python
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns

# 그래프 사이즈
plt.rcParams['figure.figsize'] = 10, 5

# 상관관계 히트맵으로 나타내기
sns.heatmap(data = cor, annot = True, fmt = '.2f', cmap = 'Blues')
plt.savefig('/content/drive/MyDrive/Colab Notebooks/github/colab/image/.wien_heatmap')
plt.show()
# annot : 수치 표시
# fmt : 소수점 자리수, '.2f' 이건 둘째 자리까지
# cmap : 히트맵 색깔, 컬러맵 인듯
```
