```python
# 레이블 인코딩
# 문자형 컬럼을 숫자로 인코딩
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()

Xtrain['Sex'] = encoder.fit_transform(Xtrain['Sex'])
Xtrain['Cabin'] = encoder.fit_transform(Xtrain['Cabin'])
Xtrain['Embarked'] = encoder.fit_transform(Xtrain['Embarked'])
Xtrain

```

```python
# 레이블 인코딩
# 문자형 컬럼을 숫자로 인코딩
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()

Xtest['Sex'] = encoder.fit_transform(Xtest['Sex'])
Xtest['Cabin'] = encoder.fit_transform(Xtest['Cabin'])
Xtest['Embarked'] = encoder.fit_transform(Xtest['Embarked'])
Xtest
```

```python
ytrain = Xtrain.pop('Survived')
ytrain
```

```python
# 학습 데이터와 테스트 데이터 분리
from sklearn.model_selection import train_test_split

Xtr, Xval, ytr, yval = train_test_split(Xtrain, ytrain, test_size=0.2, random_state = 202309)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import accuracy_score
```

```python
knnmodel = KNeighborsClassifier()
lrmodel = LogisticRegression()
dtmodel = DecisionTreeClassifier()
rfmodel = RandomForestClassifier()

# 모델별 학습 예측 평가
knnmodel.fit(Xtr, ytr)
print('KNN============')
print('Train Score : ', knnmodel.score(Xtr, ytr))
print('Test Score : ', knnmodel.score(Xval, yval))
print('정확도 : ', accuracy_score(yval, knnmodel.predict(Xval)))

lrmodel.fit(Xtr, ytr)
print('LogisticRegression============')
print('Train Score : ', lrmodel.score(Xtr, ytr))
print('Test Score : ', lrmodel.score(Xval, yval))
print('정확도 : ', accuracy_score(yval, lrmodel.predict(Xval)))

dtmodel.fit(Xtr, ytr)
print('DecisionTree============')
print('Train Score : ', dtmodel.score(Xtr, ytr))
print('Test Score : ', dtmodel.score(Xval, yval))
print('정확도 : ', accuracy_score(yval, dtmodel.predict(Xval)))

rfmodel.fit(Xtr, ytr)
print('RandomForest============')
print('Train Score : ', rfmodel.score(Xtr, ytr))
print('Test Score : ', rfmodel.score(Xval, yval))
print('정확도 : ', accuracy_score(yval, rfmodel.predict(Xval)))
```

```python
# 학습하지 않은 검증데이터로 예측
pred = rfmodel.predict(Xtest)
pred
```

```python
# 정확도 검사
accuracy_score(ytest, pred)
```
